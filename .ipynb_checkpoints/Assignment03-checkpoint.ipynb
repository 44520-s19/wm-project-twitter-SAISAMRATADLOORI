# Average number of words
Here we are calculating the average number of words for individual football players.

We will take all tweets and calculate the number of words for each tweet of an account. Then we will calculate the average of the number of words we got for each account.

```python
def word_length(text):
    tokens = text.lower().split()
    return len(tokens)

def calc_avg(words):
    return sum(words)/len(words)
```

```python
cristianoWords_Length = []
for tweet in cr7_tweets:
    cristianoWords_Length.append(word_length(tweet.text))
    
messiWords_Length = []
for tweet in leo_tweets:
    messiWords_Length.append(word_length(tweet.text))
    
ibramohivicWords_Length = []
for tweet in ibra_tweets:
    ibramohivicWords_Length.append(word_length(tweet.text))

avgOfWords = {'cristiano Tweets':calc_avg(cristianoWords_Length),'messi Tweets':calc_avg(messiWords_Length),'ibramohivic Tweets':calc_avg(ibramohivicWords_Length)}
print(avgOfWords)
```

```python
import matplotlib.pyplot as plt
heights = list(avgOfWords.values())
heights.sort(reverse=True)
plt.bar(range(len(heights)),heights)
plt.title('Average number of words ')
plt.ylabel('# number of words')
keys = list(avgOfWords.keys())
keys.sort(reverse=True, key=lambda x: avgOfWords[x])
plt.xticks(range(len(keys)),keys)
plt.show()
```
